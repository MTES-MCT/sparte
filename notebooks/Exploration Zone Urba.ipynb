```python
from public_data.models import ZoneUrba
from prettytable import PrettyTable

from django.db.models import Count
```

```python
for field in ["typezone", "origin_insee", "idurba", "lib_idzone", "partition", "destdomi"]:
    x = PrettyTable(field_names=[field, "occurences"])
    x.add_rows(
       _.values() for _ in ZoneUrba.objects.order_by(field).values(field).annotate(occurences=Count("id"))[:25]
    )
    print(f"25 premières valeurs de **{field}**")
    print(x) 
```
