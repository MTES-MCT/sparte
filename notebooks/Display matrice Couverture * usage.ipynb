```python
from prettytable import PrettyTable
import pandas as pd
from django.db.models import F
from public_data.models import CouvertureUsageMatrix


qs = CouvertureUsageMatrix.objects.exclude(label=CouvertureUsageMatrix.LabelChoices.NONE)
# qs = qs.select_related("couverture", "usage")
qs = qs.annotate(couverture_code=F("couverture__code_prefix"))
qs = qs.annotate(usage_code=F("usage__code_prefix"))
qs = qs.values("couverture_code", "usage_code", "label")
qs
```

```python
df = pd.DataFrame(list(qs))
df
```

```python
df = df.pivot(index='usage_code', columns='couverture_code', values='label')
df
```

```python
html = df.to_html(na_rep='', classes="table table-sm")
html = html.replace("<td>ARTIF_NOT_CONSU</td>", "<td bgcolor='#fc6e9e'>Artif non consommé</td>")
html = html.replace("<td>ARTIF</td>", "<td bgcolor='#fe246d'>Artificiel</td>")
html = html.replace("<td>NAF</td>", "<td bgcolor='#0e7100'>NAF</td>")
html = html.replace("<td>CONSU</td>", "<td bgcolor='#6fd261'>Consommé</td>")
print(html)
```
