```python
from public_data.models import CommunePop, Epci
from collections import defaultdict
```

```python
epci = Epci.objects.get(name__icontains="CC Coeur d'Astarac en Gascogne")
epci
```

```python
auch = epci.get_cities().filter(name="Auch").first()
auch
```

```python
from prettytable import PrettyTable
headers = ["year", "pop", "pop_change", "household", "household_change"]
x = PrettyTable(field_names=headers, align="l")
qs = CommunePop.objects.filter(city__insee="32238").order_by("year")
x.add_rows(
    [
        [getattr(c, h) for h in headers]
        for c in qs
    ]
)
x
```

```python
qs = CommunePop.objects.filter(city__in=epci.get_cities())
data = defaultdict(lambda: defaultdict(lambda: dict()))
for citypop in qs:
    if citypop.household_change:
        data[citypop.city.name][citypop.year] = citypop.household_change
data
```

```python
headers = ["city"] + [str(y) for y in range(2009, 2019)]
x = PrettyTable(field_names=headers, align="l")
x.add_rows(
    [
        [name] + [data[str(y)] for y in range(2009, 2019)]
        for name, data in data.items()
    ]
)
x
```
